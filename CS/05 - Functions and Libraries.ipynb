# Gauss

```python
from math import exp
from math import sqrt
from math import pi
from math import e
```

```python
def pdf(x):
    # by definition
    return exp(-x*x/2) / sqrt(2 * pi)

def pdfx(x, mu, sigma):
    return pdf((x - mu)/sigma) / sigma
```

```python
pdfx(10,10,10)
```

```python
def cdf(z):
    # Using Taylor series
    
    # Likelly this choise of boundary is correct to 15 places
    if z < -8: return 0
    if z > +8: return 1
    summ, term, i = 0, z ,3

    # Said to be correct to 15 places
    while summ + term != summ:
        summ += term
        term = term * z * z / i
        i += 2
    return 0.5 + summ * pdf(z)

def cdfx(z, mu, sigma):
    return cdf((z - mu) / sigma)
```

```python
# should be 0.17
cdfx(820, 1019, 209)
```

# Randomness

```python
from random import gauss
```

```python
n = 10000
x = [gauss(0.5,0.2) for _ in range(n)]
y = [gauss(0.5,0.2) for _ in range(n)]
```

```python
figure(
    data = scatter(
        x = x, 
        y = y,
        mode = "markers",
        marker = {"size": 1},
    ),
    layout = {
        "template": "plotly_white",
        "xaxis": {"visible": False},
        "yaxis": {"visible": False, "scaleanchor": "x", "scaleratio": 1},
    },    
)
```
