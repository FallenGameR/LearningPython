## LFSR

Write Linear Feed Shift Register machine and output first 2000 elements from it.
- Seed: 001000010110
- Output: 11001001011111...



```python
tap_a = 11
tap_b = 9
total = 2000
```

```python
lfsr = [0,0,1,0,0,0,0,1,0,1,1,0]
result = []

for i in range(total):
    output = lfsr[tap_a] ^ lfsr[tap_b]
    print(lfsr, "->",output)
    lfsr = [output] + lfsr[0:11]  
    result.append(output)
    
```

```python
print(*result, sep="")
```

```python
from collections import deque
lfsr = deque([0,0,1,0,0,0,0,1,0,1,1,0])

for i in range(total):
    output = lfsr[tap_a] ^ lfsr[tap_b]
    print(lfsr, "->",output)
    lfsr.pop()
    lfsr.appendleft(output)
```
